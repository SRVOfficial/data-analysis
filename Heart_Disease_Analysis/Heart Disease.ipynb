Note: Fasting blood sugar (FBS) refers to the concentration of glucose in the bloodstream after a period of fasting, typically overnight or for at least 8 hours. It is commonly used as a diagnostic tool to assess a person's blood sugar levels and screen for conditions such as diabetes mellitus.

Normal fasting blood sugar levels generally fall within the range of 70 to 100 mg/dL (3.9 to 5.6 mmol/L). Higher levels may indicate prediabetes or diabetes, while lower levels may indicate hypoglycemia (low blood sugar).

# Plot Continuous Variables

```python
data.columns
```

```python
# contains columns which has categorical values
cate_val = []

# conatains columns which has continuous values
cont_val = []


for column in data.columns:
    # The nunique() function is used to count the number of unique elements in a pandas Series or DataFrame. 
    # It returns the count of distinct observations along the specified axis.
    if data[column].nunique() <= 10:
        cate_val.append(column)
    else:
        cont_val.append(column)
```

```python
cate_val
```

```python
cont_val
```

```python
# plot continuous variables
data.hist(cont_val, figsize=(15, 7))

# to avoid overlapping
plt.tight_layout()
plt.show()
```

The plt.tight_layout() function is a method provided by Matplotlib, a popular Python plotting library. It adjusts the layout of subplots within a figure to ensure that there are no overlapping elements, making the plot more visually appealing.

When you create a plot with multiple subplots using Matplotlib, sometimes the axes labels, titles, or tick labels may overlap with each other or with neighboring subplots. In such cases, plt.tight_layout() comes in handy. It automatically adjusts the spacing between subplots to minimize overlapping and make the plot look cleane

After calling plt.tight_layout(), Matplotlib will adjust the positions of the subplots and their components (such as axes labels and titles) to improve readability and aesthetics. This function is especially useful when you have multiple subplots with different sizes or when you want to ensure that the plot looks good regardless of the size of the figure window or the aspect ratio.r.
